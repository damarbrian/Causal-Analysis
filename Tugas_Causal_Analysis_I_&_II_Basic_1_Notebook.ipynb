**Pertanyaan**
1. Jelaskan mengapa proses matching perlu dilakukan ketika ingin mengukur
dampak dari suatu treatment pada propensity score modeling?

**Jawab**
1. Proses matching dilakukan pada Propensity score modeling untuk menyeimbangkan atau menyamakan kelompok subjek penelitian sehingga dapat mengurangi bias akibat pemberian treatment yang random. Bias tersebut akan tereduksi ketika hasil perbandingan antara kelompok non-treatment dan treatment yang sedekat mungkin. Proses matching mengacu pada bangaimana menyeimbangkan data kelompok treatment dan non-treatment dengan melihat nilai yang mirip pada hasil estimasi propensity score dan melakukan pencocokan berdasarkan kovariat yang diamati, kemudian membuang varian yang tidak cocok.

**Pertanyaan**

3. Ada banyak metrik yang dapat digunakan untuk mengukur dampak dari suatu
variabel: ATE, CATE, dan LATE. Jelaskan kapan masing-masing metrik
digunakan!

**Jawab**

3. ATE merupakan ukuran yang digunakan untuk membandingkan treatment (atau intervensi) dalam eksperimen acak, evaluasi intervensi kebijakan, dan uji coba dalam aplikasi medis. ATE mengukur perbedaan hasil rata-rata (rata-rata) antara unit yang ditugaskan untuk treatment dan unit yang ditugaskan untuk kontrol. Dalam uji coba secara acak (yaitu, studi eksperimental), ATE dapat diperkirakan dari sampel menggunakan perbandingan hasil rata-rata untuk unit yang dirawat dan yang tidak diobati. Namun, ATE umumnya dipahami sebagai parameter kausal (yaitu, perkiraan atau properti populasi) yang ingin diketahui oleh peneliti, didefinisikan tanpa mengacu pada desain studi atau prosedur estimasi.

CATE merupakan ukuran yang digunakan untuk membandingkan treatment (atau intervensi) dalam eksperimen acak namun sifatnya kondisional, seperti berdasarkan variabel tertentu (gender, umur, dan lain-lain).

LATE merupakan ukuran efek treatment untuk subset dari sampel yang mengambil treatment jika dan hanya jika mereka ditugaskan untuk treatment, atau dikenal sebagai kompiler. Berbeda dengan efek treatment rata-rata (ATE), yang merupakan efek treatment tingkat subjek rata-rata; LATE seperti ATE namun memiliki kompiler. LATE dapat diestimasi dengan rasio estimasi efek intent-to-treat dan estimasi proporsi kompiler, atau alternatifnya melalui estimator variabel instrumental.

**Pertanyaan**

4. Berikut adalah pertanyaan open-ended, akan dinilai berdasarkan argumentasi
yang memiliki landasan.
Kita dapat mengekstrak knowledge dari model dengan menggunakan interpretable
machine learning. Namun kita juga dapat menggunakan metode seperti double ML
untuk mendapatkan pola sebab-akibat berdasarkan data generation process.
Mana metode yang lebih baik untuk menjelaskan dampak dari akibat? Kapan
metode interpretable machine learning dapat digunakan untuk analisis kausal?

**Jawab**

4. Metode untuk interpretasi machine learning dapat diklasifikasikan menurut berbagai kriteria. Kriteria ini membedakan apakah interpretabilitas dicapai dengan membatasi kompleksitas model machine learning (intrinsik) atau dengan menerapkan metode yang menganalisis model setelah pelatihan (post hoc). Interpretabilitas intrinsik mengacu pada model machine learning yang dianggap dapat ditafsirkan karena strukturnya yang sederhana, seperti short decision trees atau sparse linear models. Interpretasi post hoc mengacu pada penerapan metode interpretasi setelah pelatihan model. Misalnya, permutation feature importance adalah metode interpretasi post hoc. Metode post hoc juga dapat diterapkan pada interpretable models secara intrinsik. Misalnya, permutation feature importance dapat dihitung untuk decision trees. Double ML adalah kerangka kerja umum untuk memperkirakan efek kausal menggunakan teknik machine learning untuk interval confidence untuk perkiraan tersebut dan estimator yang “root n-consistent”, yaitu sebuah estimator yang memiliki properti yang baik dalam hal konvergensi dan efisiensi data.
Double ML memiliki keunggulan dalam menjelaskan hubungan dampak-akibat, karena dapat menjelaskan berbagai analisis berdasarkan treatment dan cofounding untuk melakukan inferensi kausal.
Sedangkan interpretasi machine learning untuk analisis kausal dapat digunakan dalam kondisi single analysis yang satu bentuknya adalah di mana hubungan nonkausal yang diekstraksi oleh ML yang dapat ditafsirkan digunakan untuk menyarankan hubungan sebab akibat yang potensial. Hubungan ini kemudian dapat dianalisis lebih lanjut menggunakan metode inferensi kausal dan sepenuhnya divalidasi melalui studi eksperimental menggunakan random forest atau decision trees.

Referensi: [Causal Interpretability for Machine Learning - Problems, Methods and Evaluation](https://www.researchgate.net/publication/339814440_Causal_Interpretability_for_Machine_Learning_--_Problems_Methods_and_Evaluation)
